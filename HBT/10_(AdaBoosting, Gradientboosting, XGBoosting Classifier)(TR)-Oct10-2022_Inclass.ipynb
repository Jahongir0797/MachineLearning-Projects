## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = df['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Modelling and Model Performance

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
#base_estimator=DecisionTreeRegressor(max_depth=1)
ada_model = AdaBoostClassifier(n_estimators=50, random_state=42)
```

* n_estimators: alt modellerin - week learnerlarin sayisi
* adaboost un arkasinda DT calisiyor, base estimator i da DT. 

```python
ada_model.fit(X_train,y_train)
```

```python
eval_metric(ada_model, X_train, y_train, X_test, y_test)
```

Train ve test datasi birbiriyle uyumlu, bu bizim istedigimiz bir uyum. Imbalanced bir durum olmadigini goruyoruz scorelar birbiri ile uyumlu old icin.

## Analyzing performance as more weak learners are added.

```python
error_rates = []

for n in range(1,100):
    
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    err = 1 - f1_score(y_test,preds)
    
    error_rates.append(err)
```

buyuk datada boyle bir gorsellestirme cok uzun surer, zor olur sen yapma. Datamiz kucuk old icin yapti bunu hoca. 

```python
plt.plot(range(1,100), error_rates);
```

## Gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

learning_rate ile n_estimators arasinda bir bag var. Bu ikisi uyumlu olmazsa adaboost kolaylikla overfitting e gidebilir. Zaman kaybi olmamasi adina once bu 2 parametre ile oynayip ardindan grid search e sokarsak daha az zaman harcamis oluruz ve bilgisayari daha az yorariz. Agac sayisi cok buyuk old nda (n_estimators) learning_rate i de dusuk tutarsak daha iyi sonuc veriyor.

## Comparing Models

```python
compare = pd.DataFrame({"Model": ["Logistic Regression", "KNN", "SVM", "Decision Tree", "Random Forest", "AdaBoost",
                                 "GradientBoost", "XGBoost"],
                        "F1": [log_f1, knn_f1, svc_f1, dt_f1, rf_f1, ada_f1, gb_f1, xgb_f1],
                        "Recall": [log_recall, knn_recall, svc_recall, dt_recall, rf_recall, ada_recall, gb_recall, xgb_recall],
                        "ROC_AUC": [log_auc, knn_auc, svc_auc, dt_auc, rf_auc, ada_auc, gb_auc, xgb_auc]})

def labels(ax):
    for p in ax.patches:
        width = p.get_width()                        # get bar length
        ax.text(width,                               # set the text at 1 unit right of the bar
                p.get_y() + p.get_height() / 2,      # get Y coordinate + X coordinate / 2
                '{:1.3f}'.format(width),             # set variable to display, 2 decimals
                ha = 'left',                         # horizontal alignment
                va = 'center')                       # vertical alignment
    
plt.figure(figsize=(14,10))
plt.subplot(311)
compare = compare.sort_values(by="F1", ascending=False)
ax=sns.barplot(x="F1", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(312)
compare = compare.sort_values(by="Recall", ascending=False)
ax=sns.barplot(x="Recall", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(313)
compare = compare.sort_values(by="ROC_AUC", ascending=False)
ax=sns.barplot(x="ROC_AUC", y="Model", data=compare, palette="Blues_d")
labels(ax)
plt.show()
```

Logistic Regression daha guclu bir model, en iyi F1 ve ROC_AUC skorlarina da Logistic Regression sahip. Logistic Regression'in old yerde KNN tercih edilmez. ROC_AUC bize modelin performansini gosteriyor, ayirt etme gucunu. Recall skorunda ise KNN ile LR in farki goz ardi edilebilecek duzeyde. O yuzden burada kiyasladigimizda bu datada Logistic Regression tercih edecegimiz modeldir. Eger burada XGBoost un oranlari yuksek olsaydi onu Logistic Regression'a tercih ederdik. Fakat Logistic Regression'in XGBoost'un da temelinde olan oldukca saglam ve guvenilir bir model old nu unutma. F1 ve Recall yuksekse precision zaten yuksektir, duzgundur o yuzden onu grafiklestirme geregi duymadik.
