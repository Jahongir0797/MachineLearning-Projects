#### Feature Columns
    
* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the house’s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price', axis = 1)
y = df['price']
```

```python
seed = 101
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=seed)
```

```python
from sklearn.preprocessing import MinMaxScaler  # RobustScaler()

# If there are too many outliers in the data, robust scaler should be used, otherwise minmax can be used.
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Modelling & Model Performance

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score, r2_score
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:", mse, "\nrmse:", rmse)
```

```python
X_train.shape
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(29, activation = 'relu'))
#model.add(Activation("relu")) # Activation function can be added separately as a different line after each layer.  
model.add(Dense(29, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.15, batch_size = 128, epochs = 1000)
```

```python
model.summary()
```

```python
pd.DataFrame(model.history.history)
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```

```python
model.evaluate(X_test, y_test, verbose=0)
```

```python
y_pred = model.predict(X_test)
```

```python
eval_metric(y_test, y_pred)
```

### EarlyStopping

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(29, activation = 'relu'))
model.add(Dense(29, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1))

opt = Adam(lr = 0.003)
model.compile(optimizer = opt, loss = 'mse')
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 25)
```

25 iterasyonda skor iyileşmiyorsa model.fit i durdur diyoruz patience=25 ile. 

```python
model.fit(x = X_train, y = y_train, validation_split = 0.15, batch_size = 128, epochs = 1000, callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot();
```

```python
y_pred = model.predict(X_test)
```

```python
eval_metric(y_test, y_pred)
```

küçük datasetlerinde early stopping mantıklı olabiliyor ama milyonluk data setlerinde early stopping çok tavsiye edilmez çok ilerilerde daha da düşük oranlara düşebilir loss

### Dropout

The Dropout layer randomly sets input units to 0 with a frequency of `rate`
at each step during training time, which helps prevent overfitting.

nöronların birkaçını söndürerek overfitting i engelleme yöntemi 

```python
from tensorflow.keras.layers import Dropout
```

```python
tf.random.set_seed(seed)

model = Sequential()

model.add(Dense(29, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(29, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(8, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(1))

opt = Adam(lr = 0.003)
model.compile(optimizer = opt, loss = 'mse')
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "auto", verbose = 1, patience = 25)
```

```python
model.fit(x = X_train, y = y_train, validation_split = 0.15, batch_size = 128, epochs = 1000, callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

datamız az old için dropout optimum sonuç vermedi ama datasetinin çok büyük old durumlarda dropout tavsiye edilir :) 

```python
y_pred = model.predict(X_test)
```

```python
eval_metric(y_test, y_pred)
```
