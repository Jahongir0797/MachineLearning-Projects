## Linear Regression

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):

    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},

    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls = train_val(y_train, y_train_pred, y_test, y_pred, "linear")  # modelin ad
ls
# deerler arasndaki fark ve orandan overfitting old nu gryoruz
```

## Multicollinearity - oklu Dorusallk

```python
# belirttiimiz deerden daha fazla olanlar renkli grmek iin bir fonk yazacaz

def color_red(val):

  if (0.90 < val < 1) or (-1 < val < -0.90):
    color = "red"
  elif val == 1:
    color = "blue"
  else:
    color = "white"
  return f"color: {color}" 
  # mustafa hoca deer hibir zaman -1 olmaz dedi, o yzden koda -1 i eklemedik elif ile
```

ki bamsz deiken arasnda iliki var m diye bakyoruz multicollinearity'de. nk ok ar ilikili 2 bamsz deiken modelin tahminlerine fayda salamak yerine zarar salamakta. Peki ok ar iliki kriterini neye gre belirliyoruz? Genel kabul 0.90 ve zeri ya da negatif ynde -0.90 ve zeri korelasyonun ar iliki olduu ynnde. Fakat en nihayetinde bu da bir kabul. Baz durumlarda 0.85 zerini multicollinearity kabul edebilecein gibi baz durumlarda 0.95 zerini de kabul edebilirsin. Sana kalm birazda yani.

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)
```

Modelimdeki hangi hyper parametrelerin kombinasyonu bize o model zerinde en iyi sonucu dndrecek bunu getiriyor. Btn modellerde grid search kullanabiliyoruz.
