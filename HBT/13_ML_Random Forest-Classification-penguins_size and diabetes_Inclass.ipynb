## The Data

We will be using the same dataset through our discussions on classification with tree-methods (Decision Tree,Random Forests, and Gradient Boosted Trees) in order to compare performance metrics across these related models.

We will work with the "Palmer Penguins" dataset, as it is simple enough to help us fully understand how changing hyperparameters can change classification results.

Data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network.

Gorman KB, Williams TD, Fraser WR (2014) Ecological Sexual Dimorphism and Environmental Variability within a Community of Antarctic Penguins (Genus Pygoscelis). PLoS ONE 9(3): e90081. doi:10.1371/journal.pone.0090081

Summary:
The data folder contains two CSV files. For intro courses/examples, you probably want to use the first one (penguins_size.csv).

* penguins_size.csv: Simplified data from original penguin data sets. Contains variables:

    * species: penguin species (Chinstrap, Adlie, or Gentoo)
    * culmen_length_mm: culmen length (mm)
    * culmen_depth_mm: culmen depth (mm)
    * flipper_length_mm: flipper length (mm)
    * body_mass_g: body mass (g)
    * island: island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
    * sex: penguin sex

* (Not used) penguins_lter.csv: Original combined data for 3 penguin species  

Note: The culmen is "the upper ridge of a bird's beak" 

**Our goal is to create a model that can help predict a species of a penguin based on physical attributes, then we can use that model to help researchers classify penguins in the field, instead of needing an experienced biologist**

## Ordinal Encoder

https://bookdown.org/max/FES/categorical-trees.html

https://towardsdatascience.com/one-hot-encoding-is-making-your-tree-based-ensembles-worse-heres-why-d64b282b5769

* dummies numeric datalarda uygulanabilir. Tree based modeller disinda dummies uygulariz, ama tree based modellerde kesinlikle yapilmasi gereken encoderdir. Dummies de uygulanabilir ama efektif olmaz. 
* Tree based modeller kategorik datalarda modele sordugu sorunun cevabina gore kategorik bir bolunme/tahmin yaparken; numeric datalarda ise her observation da ayni soruyu soruyor, buradaki degerin altini ve ustunu sorguluyor, gini kat sayisinin en dusuk old feature i best feature olarak aliyor, o gini katsayisina ait score un altini ve ustunu sorgulayarak leaf node lara boluyor. 
* Kategorik datalarda eger hiyerarsik bir durum yoksa onu dummies feature a donusturuyoruz, ancak linear modellerde arka planda bir istatistik islemi yapildigindan encoder yapamiyoruz. YANI:
* Istatistiksel bir islem yapildiginda encoder dogru score vermeyecek. O yuzden encoder Tree Based Modellerde uygulanmasi acisindan gayet basarili, complexity i de azaltiyor. 
* Dummies yaptigimizda complexitymiz artiyor. Tree based modellerde dummies uygulanirsa scorelar dogru ya da yaklasik ciksa bile dummies feature importance in yanlis belirlenmesine sebep oluyor. 
* Regresyon modellerinde hiyerarsik bir durum olmadiginda label encoder yapsak degerler uzerinde islem yapilacagindan skorlarimiz hatali gelir.
* Tree based modellerde lamparametric oldndan sadece if else seklinde ilerledigi icin burada encoder yapabiliyoruz, hatta encoder yapmak scorelar acisindan daha iyidir tree based modellerde. 
* Feature Importance icin label encoder yapmamiz daha dogru sonuc verir. 
* 2 cesit encoder var:
> 1. Label Encoder: Featurelari tek tek yakaliyorsunuz, yani tek tek uygulayabiliyorsunuz featurelara. 
> 2. Ordinal Encoder: Kodu yazdiktan sonra butun kategorik olan featurelara otomatik olarak bunu uyguluyor. Bu yuzden kullanimi daha kolay, cogunlukla ordinal encoder i kullanacagiz.

## Modelling and Model Performance

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix

def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

Random Forest'daki parametreler:
* n_estimators: olusturacagi agac sayisi
* max_features: feature sayisinin karekokunu aliyor, featurelardan en iyi olanindan baslamasini sagliyor. defaultu sqrt ama log2, None gibi baska secenekleri de var. 
> n_estimators ve max_features overfitting ile mucadele de en onemli iki parametremiz. 
Mesela max_features=3 old nda 9 feature li bir veri de random forest da rastgele 3 tanesini secip aliyor onlardan en iyisini aliyor sonra baska bir rastgele 3lu alip bu islemleri bu sekilde devam ettiriyor, random olma durumu cokca old icin overfittingin de onune gecilmis oluyor. 
* bootstrap: datanin modeli olustururken subsample dedigimiz 2/3 u ile calisiyor. O gozlemler uzerinden datamin gozlemlerini max_features uzerinden yeniden olusturuyor, oransallik sayesinde modelin daha iyi genellemeler yaparak daha iyi tahminler almasini sagliyor. Default degeri True. Random Forest'i Random yapan parametredir.
* class_weight: default u None. 'balanced', 'balanced_subsampled' olarak da ayarlayabiliyoruz. Istersek logistic reggression daki balanced isleminde de uygulayabiliyoruz bunu. 
* max_samples: 'None' dersek ki default degeri None, butun datadaki verileri aliyor. Ama 0.8 dersek mesela datanin %80'ini aliyor, sonra Random Forest bu %80'in 2/3 unu aliyor.
